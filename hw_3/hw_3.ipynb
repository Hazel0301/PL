# get the title and url

```python
title = soup.title
print(title.text)
```

```python
#article info
articles = soup.find_all('div', class_='r-ent')

# every article
for article in articles:
    
    date = article.find('div', class_='date').text.strip()   
    
    title_element = article.find('div', class_='title') #title and url
    if title_element is not None:
        title = title_element.text.strip()
        url = 'https://www.ptt.cc' + title_element.find('a')['href']
    else:
        title = 'No title'
        url = ''

    # print
    print(date,title,"\n", url)
```

# Classify the articles by group
# -->NMIXX, SK, JYPE, NiziU

```python
categories = {
    'NMIXX': [],
    'SK': [],
    'JYPE': [],
    'NiziU': []
}

url = 'https://www.ptt.cc/bbs/JYPnation/index481.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
titles = soup.select('.title')

for title in titles:
    text = title.text
    if '[NMIXX ]' in text:
        categories['NMIXX'].append(text)
    elif '[  SK  ]' in text:
        categories['SK'].append(text)
    elif '[ JYPE ]' in text:
        categories['JYPE'].append(text)
    elif '[NiziU ]' in text:
        categories['NiziU'].append(text)

for category, titles in categories.items():
    print(f'{category}: {len(titles)}')
    articles = soup.find_all('div', class_='r-ent')
    for article in articles:
        date = article.find('div', class_='date').text.strip()  
        
        title_element = article.find('div', class_='title') #title and url
        title = title_element.text.strip()
        url = 'https://www.ptt.cc' + title_element.find('a')['href']
        
    for title in titles:
        print(date,title, url)
```

# get the data from every url

```python
url_list = []

for a in articles:
    title_element = a.find('div', class_='title') #title and url
    if title_element is not None:
        url_list.append("https://www.ptt.cc" + title_element.find('a')['href'])
    else:
        url_list.append("")
```

```python
# track data in every url
for url in url_list:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # catch every classification of data
    content = ""
    for tag in soup.select(".article-metaline, .article-metaline-right, .push, .f2"):
        if tag.name == "div" and "class" in tag.attrs and "richcontent" in tag.attrs["class"]:
            continue  # 忽略影片嵌入內容
        content += tag.text + "\n"
        
    print("內文：", content) #輸出
    print("--------------------------------------------------")
```

# save data as csv form

```python
import csv

with open('JYP.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['date', 'title', 'url'])

    for article in articles:
        date = article.find('div', class_='date').text.strip()
        title_element = article.find('div', class_='title')

        if title_element is not None:
            title = title_element.text.strip()
            url = 'https://www.ptt.cc' + title_element.find('a')['href']
        else:
            title = 'No title'
            url = ''

        writer.writerow([date, title, url])

test = pd.read_csv("JYP.csv")
test
```

# save data as json form

```python
JYP_data = []
with open('JYP.csv', newline='', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        JYP_data.append(row)

with open('JYP.json', 'w', encoding='utf-8') as file:
    json.dump(JYP_data, file, ensure_ascii=False)
    
with open('JYP.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
data
```
