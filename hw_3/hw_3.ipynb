### response = 200 --> url available

```python
web = requests.get(url) #ewb page info
soup = BeautifulSoup(web.text, "html.parser")
```

```python
title = soup.title
print(title.text)
```

```python
#article info
articles = soup.find_all('div', class_='r-ent')

# every article
for article in articles:
    
    date = article.find('div', class_='date').text.strip()   
    
    title_element = article.find('div', class_='title') #title and url
    if title_element is not None:
        title = title_element.text.strip()
        url = 'https://www.ptt.cc' + title_element.find('a')['href']
    else:
        title = 'No title'
        url = ''

    # print
    print(date,title,"\n", url)
```

# save data as csv form

```python
import csv

with open('JYP.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['date', 'title', 'url'])

    for article in articles:
        date = article.find('div', class_='date').text.strip()
        title_element = article.find('div', class_='title')

        if title_element is not None:
            title = title_element.text.strip()
            url = 'https://www.ptt.cc' + title_element.find('a')['href']
        else:
            title = 'No title'
            url = ''

        writer.writerow([date, title, url])

test = pd.read_csv("JYP.csv")
test
```

# save data as json form

```python
JYP_data = []
with open('JYP.csv', newline='', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        JYP_data.append(row)

with open('JYP.json', 'w', encoding='utf-8') as file:
    json.dump(JYP_data, file, ensure_ascii=False)
    
with open('JYP.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
data
```
